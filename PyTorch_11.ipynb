# Image Filters and Kernels Overview

This notebook is a simple beginner introduction to image filters (kernels) for CNNs.

## 1) What Is an Image Filter (Kernel)?

In CNNs, image filter and kernel mean the same thing.

A kernel is a small matrix (often 3 x 3) that moves across an image and performs calculations to detect patterns.

## 2) Why Kernels Matter in CNNs

The first major CNN step is convolution.

During convolution, the kernel helps the network find useful visual features, such as:
- edges
- outlines
- directional patterns (left-right or top-bottom changes)

## 3) How Convolution Works (Simple View)

Think of an image as a grid of numbers (pixel intensities).

To apply convolution:
1. Place a 3 x 3 kernel on the top-left part of the image.
2. Multiply each kernel value by the matching image value.
3. Add all 9 results into one number.
4. Move the kernel and repeat.

This repeated process builds a new feature map.

## 4) What Is Stride?

Stride is how far the kernel moves each step.

- Stride 1: move one pixel at a time
- Stride 2: move two pixels at a time

Larger stride means fewer convolution positions and a smaller output map.

## 5) Fixed Filters vs Learned Filters

Traditional image processing uses hand-designed filters (for example blur or sharpen).

In CNNs, filter values are learned automatically during training.

That means the model discovers which kernel values are best for the task.

## 6) Edge Detection Idea (Sobel Family)

Some kernels are known for edge detection, such as Sobel-style filters.

Different orientations emphasize different edges:
- left or right boundaries
- top or bottom boundaries

This is useful because object shape often depends on edges and contours.

## 7) CNN Goal in This Context

For handwritten digit recognition, convolution helps highlight digit structure against background.

After useful features are extracted, later layers classify the image as one of the digits 0 to 9.

## 8) Key Takeaways

- Kernel and filter are interchangeable terms.
- Convolution means sliding a small matrix over the image and computing multiply-plus-sum values.
- Stride controls step size while sliding.
- CNNs learn filter values automatically from data.
- Edge and contour information is a core reason CNNs work well for images.
