# Imports

```python
import torch # Import the torch library for tensor operations
```

# Reshape & View

```python
rand_tensor = torch.randint(1, 13, [1, 12]) # Create a 1D tensor with 12 random integers between 1 and 12
print(rand_tensor) # Print the generated tensor
```

```python
tensor_2d_reshape = rand_tensor.reshape(-1, 3) # Reshape the 1D tensor into a 2D tensor with 3 columns, inferring the number of rows
print(tensor_2d_reshape) # Print the reshaped tensor
print(tensor_2d_reshape.shape) # Print the shape of the reshaped tensor
```

```python
tensor_2d_view = rand_tensor.view(-1, 3) # Create a view of the 1D tensor as a 2D tensor with 3 columns, inferring the number of rows
print(tensor_2d_view) # Print the viewed tensor
print(tensor_2d_view.shape) # Print the shape of the viewed tensor
```

# Slicing

```python
random_tensor = torch.rand((4, 5)) + 1 # Create a 4x5 tensor with random floats between 1 and 2
print(random_tensor) # Print the generated tensor
```

```python
print(random_tensor[:2, :]) # Print the first two rows and all columns of the tensor
```

```python
print(random_tensor[3:, :2]) # Print the last row and the first two columns of the tensor
```
