# Convolutional Neural Network (CNN) Intro

This is a simple beginner-friendly introduction to CNNs.

## 1) What Is a CNN?

A CNN (Convolutional Neural Network) is a type of neural network made for image tasks.
It learns patterns like edges and shapes, then predicts what the image contains.

## 2) Why CNN Instead of a Basic Neural Network?

Basic neural networks can work, but images have structure (left-right, top-bottom).
CNNs are better at keeping and learning from that structure.

## 3) MNIST Dataset (Used in This Lesson)

MNIST contains handwritten digits from 0 to 9.
- Image size: 28 x 28
- Grayscale images
- 60,000 training images
- 10,000 testing images

MNIST is clean and labeled, so it is excellent for beginners.

## 4) How the Computer Sees a Digit

An image is represented as a grid of numbers.
- Values near 0 are usually background
- Values near 1 are darker parts of the digit
- Values in between represent softer edges

This helps the model find borders and important parts of the digit.

## 5) CNN Flow (High Level)

1. Convolution: detects useful features (such as edges).
2. Pooling: shrinks the data while keeping important information.
3. Flatten: converts feature maps into a list of numbers.
4. Dense layers: make the final prediction (digit 0 to 9).

## 6) Training vs Testing

- Training data teaches the model.
- Testing data checks performance on unseen data.

Good test performance means the model learned real patterns.

## Key Takeaways

- CNNs are powerful for image classification.
- MNIST is a great beginner dataset.
- Pixel values help the model detect shape and border information.
- The model finally classifies the image into one of 10 classes.
